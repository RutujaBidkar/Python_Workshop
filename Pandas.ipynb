```python
df["promoted Salary"] = df["Salary"] *10
df
```

```python
#Data Cleaning
df.isnull().sum()
```

```python
df.dropna()           #delete Null value
```

```python
df.dropna(how = "any")      #any row that had any null value
```

```python
df.dropna(how = "all") #if all the values in any row are null
```

```python
df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"].fillna(df["Salary"].median())
```

```python
#forward Fill : Move from Top to bottom
#backward Fill : move from bottom to top
```

```python
df["Age"].fillna(method = "ffill")
```

```python
df["Age"].fillna(method = "bfill")
```

```python
df["Name"].replace("Charlie","Rose")
```

```python
df
```

```python
#duplicate values
```

```python
df_dup = df[df.duplicated(keep ="first")]
df_dup
```

```python
df = df.drop_duplicates()
df
```

```python
#Invaild values :
#lambda -> python
df["promoted Salary"] = df['promoted Salary'].apply(lambda x: x/10 if x > 650000 else x)
```

```python
df
```

```python
name = "Alice_ fernandes"
df[["first_name","Last_name"]] = df["name"].str.split("_")
```

```python
# apply and lambda
def multipliying_age(x):
  return x*2

df["Age"]= df["Age"].apply(multipliying_age)
```

```python
df["Age"] = df["Age"].apply(lambda x:x/2)
df
```
