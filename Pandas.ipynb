```python
fruit_protein ={
    "Avocado" : 2.0,
    "guva" : 2.6,
    "Blackberries": 2.0,
    "Oranges" : 0.9,
    "Banana" : 1.1,
    "Apples" : 0.3,
    "Kiwi": 1.1,
    "Pomegranate" : 1.7,
    "Mango" : 0.8,
    "cherries" : 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name = "Protein")
s2
```

```python
#Conditional Selection
```

```python
s2[s2>1]
```

```python
#Logical Operators and or not
s2[(s2>0.5) & (s2<=2)]
```

```python
s2[(s2>0.5) | (s2<=2)]
```

```python
#not operation:
s2[~(s2>1)]
```

```python
#Modyfying the series:
s2["Mango"] = 2.8
s2
```
